```python
import numpy as np
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from pandas import DataFrame
X, _ = make_blobs(n_samples = 500, centers = 3, n_features = 2, random_state = 20)
```

```python
X[0]
```

```python
df = DataFrame(dict(x=X[:,0], y=X[:,1]))
fig, ax = plt.subplots(figsize=(8,8))
df.plot(ax=ax, kind='scatter', x='x', y='y')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps = 1, min_samples = 5).fit(X)
cluster = clustering.labels_
cluster
```

```python
#length of cluster gives an idea of outliers
len(set(cluster))
```

```python
#We saw that our dataset contains three clusters.
#The first cluster that is the -1 level represents the noise in this DBSCAN.
```

```python
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=cluster))
colors = {-1: 'red', 0: 'blue', 1:'orange', 2:'green', 3:'yellow'}
fig, ax = plt.subplots(figsize=(8,8))
grouped = df.groupby('label')
for key, group in grouped:
  group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

```python
#Task:
#Use https://www.geeksforgeeks.org/ml-k-means-algorithm/ and https://www.geeksforgeeks.org/ml-mini-batch-k-means-clustering-algorithm/
#to practice labs for k-means++ and mini-batch k-means
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/driver-data.csv')
```

```python
print(df.columns)
```

```python
# Feature selection
X = df[['mean_dist_day', 'mean_over_speed_perc']]
```

```python
# Standardize the features
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-means++ clustering
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans.fit(X_scaled)
```

```python
# Mini-Batch K-means clustering
mini_kmeans = MiniBatchKMeans(n_clusters=3, init='k-means++', random_state=42, batch_size=100)
mini_kmeans.fit(X_scaled)
```

```python
# Plotting the results
plt.figure(figsize=(14, 7))
```

```python
# K-means++ result
plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', marker='o')
plt.title('K-means++ Clustering')
plt.xlabel('Mean Distance per Day')
plt.ylabel('Mean Over Speed Percentage')
```

```python
# Mini-Batch K-means result
plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=mini_kmeans.labels_, cmap='viridis', marker='o')
plt.title('Mini-Batch K-means Clustering')
plt.xlabel('Mean Distance per Day')
plt.ylabel('Mean Over Speed Percentage')
plt.show()
```
